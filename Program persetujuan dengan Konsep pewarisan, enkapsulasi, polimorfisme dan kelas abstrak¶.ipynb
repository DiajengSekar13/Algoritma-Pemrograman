Konsep pewarisan dalam pemrograman berorientasi objek adalah sebuah mekanisme di mana sebuah kelas dapat mewarisi atribut dan metode dari kelas induk atau super class-nya. Dengan kata lain, pewarisan memungkinkan kelas turunan atau sub class untuk menggunakan kembali kode yang ada pada kelas induknya tanpa perlu menuliskan kembali kode tersebut.

1. Class Orang merupakan class dasar yang memiliki dua method yaitu setuju_proposal() dan tolak_proposal().
2. Class Mahasiswa merupakan turunan dari class Orang dan memiliki method submit_proposal() dan get_proposal() yang terkait dengan pengajuan dan pengambilan proposal skripsi.
3. Class Dosen merupakan turunan dari class Orang dan memiliki method setuju_proposal() dan tolak_proposal() yang berperan dalam menyetujui atau menolak proposal skripsi yang diajukan oleh mahasiswa.
4. Class Pembimbing1 dan Pembimbing2 merupakan turunan dari class Dosen dan masing-masing memiliki method setuju_proposal() dan tolak_proposal() yang berperan sebagai pembimbing 1 dan pembimbing 2 pada proposal skripsi.
5. Class ProposalSkripsi berperan sebagai representasi dari sebuah proposal skripsi dan memiliki beberapa atribut seperti judul, mahasiswa, abstrak, dosen pembimbing 1 dan 2, serta status proposal. Class ini juga memiliki beberapa method yang berperan dalam pengaturan informasi yang terkait dengan proposal skripsi.

### Konsep Enkapsulasi

```python
class Mahasiswa:
    def __init__(self, nama, nim, angkatan):
        self.nama = nama
        self.nim = nim
        self.angkatan = angkatan
        self.proposal = None
    
    def submitProposal(self, judul, abstrak):
        self.proposal = ProposalSkripsi(judul, self.nama, abstrak)
    
    def getProposal(self):
        return self.proposal


class Dosen:
    def __init__(self, nama, nip, jabatan):
        self.nama = nama
        self.nip = nip
        self.jabatan = jabatan
    
    def setujuProposal(self, proposal):
        proposal.setDosen(self.nama)
        proposal.setStat("Disetujui")
    
    def tolakProposal(self, proposal):
        proposal.setDosen(self.nama)
        proposal.setStat("Ditolak")


class ProposalSkripsi:
    def __init__(self, judul, mahasiswa, abstrak):
        self.__judul = judul
        self.__mahasiswa = mahasiswa
        self.__abstrak = abstrak
        self.__dosen1 = None
        self.__dosen2 = None
        self.__status = "Belum disetujui"
    
    def setDosen(self, dosen):
        if self.__dosen1 is None:
            self.__dosen1 = dosen
        elif self.__dosen2 is None:
            self.__dosen2 = dosen
    
    def setStat(self, status):
        self.__status = status
    
    def getInfo(self):
        info = "Judul: " + self.__judul + "\n"
        info += "Mahasiswa: " + self.__mahasiswa + "\n"
        info += "Dosen Pembimbing 1: " + self.__dosen1 + "\n"
        info += "Dosen Pembimbing 2: " + self.__dosen2 + "\n"
        info += "Abstrak: " + self.__abstrak + "\n"
        info += "Status: " + self.__status + "\n"
        return info
```

```python
mhs1 = Mahasiswa("Dicky Maulana", "22083010003", "2018")
dosen1 = Dosen("Dr. Siti Aisyah", "22870", "Lektor")
dosen2 = Dosen("Prof. Joko Samudera", "22897", "Profesor")
mhs1.submitProposal("Dampak Keunikan Produk Dalam Negeri Terhadap Daya Tarik Wisatawan Mancanegara", "Abstrak dari proposal skripsi")
dosen1.setujuProposal(mhs1.getProposal())
dosen2.setujuProposal(mhs1.getProposal())
print(mhs1.getProposal().getInfo())
```

### Konsep Polimorfisme

```python
class Manusia:
    def __init__(self, nama):
        self.nama = nama

    def persetujuan(self, proposal):
        pass


class Dosen(Manusia):
    def persetujuan(self, proposal):
        proposal.setDosen1(self.nama)
        proposal.setStat("Disetujui oleh Dosen")


class Pembimbing1(Manusia):
    def persetujuan(self, proposal):
        proposal.setDosen1(self.nama)
        proposal.setStat("Disetujui oleh Pembimbing 1")


class Pembimbing2(Manusia):
    def persetujuan(self, proposal):
        proposal.setDosen2(self.nama)
        proposal.setStat("Disetujui oleh Pembimbing 2")


class Mahasiswa:
    def __init__(self, nama, nim, angkatan):
        self.nama = nama
        self.nim = nim
        self.angkatan = angkatan
        self.proposal = None

    def submitProposal(self, judul, abstrak):
        self.proposal = ProposalSkripsi(judul, self.nama, abstrak)

    def getProposal(self):
        return self.proposal

    def persetujuan(self, proposal, obj):
        obj.persetujuan(proposal)


class ProposalSkripsi:
    def __init__(self, judul, mahasiswa, abstrak):
        self.__judul = judul
        self.__mahasiswa = mahasiswa
        self.__abstrak = abstrak
        self.__dosen1 = None
        self.__dosen2 = None
        self.__status = "Belum disetujui"

    def setDosen1(self, dosen):
        self.__dosen1 = dosen

    def setDosen2(self, dosen):
        self.__dosen2 = dosen

    def setStat(self, status):
        self.__status = status

    def getInfo(self):
        info = "Judul: " + self.__judul + "\n"
        info += "Mahasiswa: " + self.__mahasiswa + "\n"
        info += "Dosen Pembimbing 1: " + str(self.__dosen1) + "\n"
        info += "Dosen Pembimbing 2: " + str(self.__dosen2) + "\n"
        info += "Abstrak: " + self.__abstrak + "\n"
        info += "Status: " + self.__status + "\n"
        return info
```

```python
mahasiswa1 = Mahasiswa("Dicky Maulana", "18083010006", "2018")
dosen1 = Dosen("Dr. Gatot Katcha")
pembimbing1 = Pembimbing1("Dr. Siti Aisyah")
pembimbing2 = Pembimbing2("Prof. Djoko Samudro")
mahasiswa1.submitProposal("Dampak Keunikan Produk Dalam Negeri Terhadap Daya Tarik Wisatawan Mancanegara", "Penelitian ini bertujuan untuk mengoptimalkan proses produksi dengan menggunakan metode penelitian Kuantitatif.")
proposal1 = mahasiswa1.getProposal()
mahasiswa1.persetujuan(proposal1, dosen1)
mahasiswa1.persetujuan(proposal1, pembimbing1)
mahasiswa1.persetujuan(proposal1, pembimbing2)

print(proposal1.getInfo())
```

Kelas Mahasiswa berfungsi untuk merepresentasikan objek mahasiswa yang memiliki atribut nama, nim, angkatan, dan proposal. Kelas ini memiliki method submitProposal() untuk mengajukan proposal skripsi dan getProposal() untuk mengambil proposal yang telah diajukan.

Kelas Dosen merupakan superclass yang merepresentasikan objek dosen dengan atribut nama, nip, dan jabatan. Kelas ini memiliki method setujuProposal() untuk menyetujui proposal skripsi dan tolakProposal() untuk menolak proposal skripsi.

Kelas Pembimbing1 dan Pembimbing2 merupakan subclass dari kelas Dosen yang mewarisi atribut dan method dari superclass-nya. Kedua kelas ini memiliki method setujuProposal() yang memperbaharui atribut dosen pada proposal skripsi yang bersangkutan.

Kelas ProposalSkripsi merepresentasikan objek proposal skripsi dengan atribut judul, mahasiswa, abstrak, dosen1, dosen2, dan status. Kelas ini memiliki method setDosen1(), setDosen2(), setStat(), dan getInfo() yang masing-masing berguna untuk memperbaharui atribut dosen, status, dan mengambil informasi terkait proposal skripsi.

Dari penjelasan tersebut, dapat disimpulkan bahwa kode di atas menggunakan konsep pewarisan, enkapsulasi, dan polimorfisme. Konsep pewarisan terlihat pada kelas Pembimbing1 dan Pembimbing2 yang mewarisi atribut dan method dari kelas Dosen. Konsep enkapsulasi terlihat pada penggunaan atribut dengan access modifier private (diawali dengan __). Konsep polimorfisme terlihat pada penggunaan method setujuProposal() yang memiliki implementasi yang berbeda pada kelas Dosen, Pembimbing1, dan Pembimbing2.

### Konsep Abstrak

```python
from abc import ABC, abstractmethod

class Orang(ABC):
    def __init__(self, nama):
        self.nama = nama
    
    @abstractmethod
    def setujuProposal(self, proposal):
        pass
    
    @abstractmethod
    def tolakProposal(self, proposal):
        pass


class Mahasiswa:
    def __init__(self, nama, nim, angkatan):
        self.nama = nama
        self.nim = nim
        self.angkatan = angkatan
        self.proposal = None
    
    def submitProposal(self, judul, abstrak):
        self.proposal = ProposalSkripsi(judul, self.nama, abstrak)
    
    def getProposal(self):
        return self.proposal


class Dosen(Orang):
    def __init__(self, nama, nip, jabatan):
        super().__init__(nama)
        self.nip = nip
        self.jabatan = jabatan
    
    def setujuProposal(self, proposal):
        proposal.setDosen1(self.nama)
        proposal.setStat("Disetujui")
    
    def tolakProposal(self, proposal):
        proposal.setDosen1(self.nama)
        proposal.setStat("Ditolak")


class Pembimbing1(Orang):
    def __init__(self, nama, nip, jabatan):
        super().__init__(nama)
        self.nip = nip
        self.jabatan = jabatan
    
    def setujuProposal(self, proposal):
        proposal.setDosen1(self.nama)
        proposal.setStat("Disetujui oleh Pembimbing 1")
    
    def tolakProposal(self, proposal):
        proposal.setDosen1(self.nama)
        proposal.setStat("Ditolak oleh Pembimbing 1")


class Pembimbing2(Orang):
    def __init__(self, nama, nip, jabatan):
        super().__init__(nama)
        self.nip = nip
        self.jabatan = jabatan
    
    def setujuProposal(self, proposal):
        proposal.setDosen2(self.nama)
        proposal.setStat("Disetujui oleh Pembimbing 2")
    
    def tolakProposal(self, proposal):
        proposal.setDosen2(self.nama)
        proposal.setStat("Ditolak oleh Pembimbing 2")


class ProposalSkripsi:
    def __init__(self, judul, mahasiswa, abstrak):
        self.__judul = judul
        self.__mahasiswa = mahasiswa
        self.__abstrak = abstrak
        self.__dosen1 = None
        self.__dosen2 = None
        self.__status = "Belum disetujui"
    
    def setDosen1(self, dosen):
        self.__dosen1 = dosen
    
    def setDosen2(self, dosen):
        self.__dosen2 = dosen
    
    def setStat(self, status):
        self.__status = status
    
    def getInfo(self):
        info = "Judul: " + self.__judul + "\n"
        info += "Mahasiswa: " + self.__mahasiswa + "\n"
        info += "Dosen Pembimbing 1: " + str(self.__dosen1) + "\n"
        info += "Dosen Pembimbing 2: " + str(self.__dosen2) + "\n"
        info += "Abstrak: " + self.__abstrak + "\n"
        info += "Status: " + self.__status + "\n"
        return info
```

```python
dosen1 = Dosen("Dr. Gatot Katcha", "23450", "Guru Besar")
pembimbing1 = Pembimbing1("Dr. Siti Aisyah", "22870", "Lektor")
pembimbing2 = Pembimbing2("Prof. Djoko Samudro", "1119", "Dosen Tetap")

mahasiswa1 = Mahasiswa("Dicky Maulana", "18083010006", "2018")
mahasiswa1.submitProposal(" Dampak Keunikan Produk Dalam Negeri Terhadap Daya Tarik Wisatawan Mancanegara", "Penelitian ini bertujuan untuk mengoptimalkan proses produksi dengan menggunakan metode penelitian Kuantitatif.")

pembimbing1.setujuProposal(mahasiswa1.getProposal())
pembimbing2.setujuProposal(mahasiswa1.getProposal())
dosen1.setujuProposal(mahasiswa1.getProposal())
print(mahasiswa1.getProposal().getInfo())
```
